## Problem 1: Check if a word starts with a capital letter

```python
sample = "Hello World"
m = re.match(r"[A-Z][a-z]*", sample)

print("Match: ", m.group(0) if m else "No match")
```

## Problem 2: Extract URLs

```python
sample = "Visit https://google.com or http://example.org for more information."

urls = re.findall(r"https?://[A-Za-z0-9./]+", sample)

print("Extracted URLs:", urls)
```

## Problem 3: Remove Extra Spaces

```python
sample = "this  is  NLP   lab."

cleaned_text = re.sub(r"\s+", " ", sample)

print("Cleaned Text:", cleaned_text)
```

## Problem 4: Mask Email Username

```python
email = "test123@gmail.com"
masked = re.sub(r"^[^@]+", "****", email)

print("Masked Email:", masked)
```

## Problem 5: Validate Strong Password (Without Lookaheads)
conditions:
1. at least 8 characters
2. one uppercase, one lowercase
3. one number, one special character

```python
password = "Abc@1234"

pattern = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$"

print("Strong Password" if re.match(pattern, password) else "Weak Password")
```

## Problem 6: Extract Hashtags

```python
sample = "#NLP is #amazing in #AI_2025 !"

hashtags = re.findall(r"#\w+", sample)

print("Extracted Hashtags:", hashtags)
```

## Problem 7: Extract Strong Password with Positive Lookaheads

```python
password = "Abc@1234"

pattern = r"^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@#$%&]).{8,}$"

print("Strong Password" if re.match(pattern, password) else "Weak Password")
```

## Problem 8: Detect Repeated Words using backreferences

```python
sentence = "This is a test test sequence"

pattern = r"\b(\w+)\s+\1\b"

repeated_words = re.findall(pattern, sentence)

print("Repeated Words:", repeated_words)
```
